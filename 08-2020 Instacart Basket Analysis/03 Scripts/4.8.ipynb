#### prodcut_name is Lowfat 2% Milkfat Cottage Cheese and 2 % Reduced Fat Milk

```python
dairy = df_orders_products_merged.loc[df_orders_products_merged['department_id'].isin([16])]
dairy_without_weird = dairy.loc[~dairy['prices'].isin([99999, 14900])]
print('Filter for records with the department_id value of 16, while excluding records without the unusually high prices:')
dairy_without_weird.sample(5)
```

```python
print('Maximum price in the dairy_without_weird filter:')
dairy_without_weird['prices'].max()
```

```python
reduced_fat_milk = dairy_without_weird.query('product_name.str.contains("Reduced Fat") & product_name.str.contains("2") & product_name.str.contains("Milk")')
```

```python
print('Sample output from reduced_fat_milk:')
reduced_fat_milk.sample(10)
```

```python
print('Quick statistics in reduced_fat_milk:')
reduced_fat_milk.describe()
```

#### mean, min and max of anything with reduced fat 2 and milk are 9.08, 1.2, and 14.7. with that making more sense we can change product_id 33664 to 9.99 

```python
cottage_cheese = dairy_without_weird.query('product_name.str.contains("Cottage Cheese") & product_name.str.contains("2") & product_name.str.contains("Milkfat")')
```

```python
print('Sample output from cottage_cheese:')
cottage_cheese.sample(10)
```

```python
print('Quick statistics in cottage_cheese:')
cottage_cheese.describe()
```

##### mean, min, and max for things containing cottage cheese, 2 and milkfat are 7.23, 2.4, and 11.8. with that making more sense we can change product_id 21553 to reflect the mean value.

```python
df_orders_products_merged['prices'] = df_orders_products_merged['prices'].replace({99999: 9.99, 14900: cottage_cheese['prices'].mean()})
```

```python
print('The statistics of the new prices column in df_orders_products_merged:')
df_orders_products_merged['prices'].describe()
```

```python
print('The mean, min and max of prices grouped by loyalty_flag with the updated prices column in df_orders_products_merged:')
df_orders_products_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

### new customer now holds highest average purchases. max value is more realistic.

```python
df_orders_products_merged['average_price'] = df_orders_products_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_orders_products_merged['average_price'] = df_orders_products_merged['average_price'].round(2)
```

```python
df_orders_products_merged.loc[df_orders_products_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_orders_products_merged.loc[df_orders_products_merged['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_orders_products_merged['median_prior_orders'] = df_orders_products_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_orders_products_merged.loc[df_orders_products_merged['median_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
